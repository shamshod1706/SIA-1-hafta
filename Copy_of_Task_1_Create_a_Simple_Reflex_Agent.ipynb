# **1- MASALA:** **Vacuum - Cleaner World agentini yaratish. (2 ta xona uchun)**



---
##**Maqsad:**
##2 ta xonali (A va B) muhitda tozalikni saqlash.

## **Vazifa:**

## Agent hozir qaysi xonada ekanligini va xonaning holatini (toza yoki kir) aniqlashi kerak.

* ## *Agar xona "Kir" bo'lsa -> "Tozalamoq".*

* ## *Agar xona "Toza" bo'lsa va A da bo'lsa -> "O'ngga yurish".*

* ## *Agar xona "Toza" bo'lsa va B da bo'lsa -> "Chapga yurish".*


#**‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>xonalar = {"A": "iflos", "B": "toza"}</code>

* ### <code>agent_joylashuvi = "A"</code>

##***Klasslar:***

* ### <code>Muhit</code>
  * ### <code>perseptsiya_ol()</code>
  * ### <code>harakat_qollash(harakat)</code>
* ### <code>RefleksAgent</code>
  * ### <code>qaror_qabul_qil(perseptsiya)</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas
 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.

```python
# Kodni shu yerga yozing

class Muhit:
    def __init__(self):
        self.xonalar = {"A": "iflos", "B": "toza"}
        self.agent_joylashuvi = "A"

    def perseptsiya_ol(self):
        holat = self.xonalar[self.agent_joylashuvi]
        return (self.agent_joylashuvi, holat)

    def harakat_qollash(self, harakat):

        if harakat == "Tozalamoq":
            self.xonalar[self.agent_joylashuvi] = "toza"

        elif harakat == "O'ngga yurish":
            self.agent_joylashuvi = "B"

        elif harakat == "Chapga yurish":
            self.agent_joylashuvi = "A"


class RefleksAgent:

    def qaror_qabul_qil(self, perseptsiya):
        joy, holat = perseptsiya

        if holat == "iflos":
            return "Tozalamoq"

        elif holat == "toza" and joy == "A":
            return "O'ngga yurish"

        elif holat == "toza" and joy == "B":
            return "Chapga yurish"

def simulyatsiya(qadamlar=20):
    muhit = Muhit()
    agent = RefleksAgent()

    for i in range(1, qadamlar + 1):
        perseptsiya = muhit.perseptsiya_ol()
        harakat = agent.qaror_qabul_qil(perseptsiya)

        print(f"Qadam {i}")
        print("Percept:", perseptsiya)
        print("Action:", harakat)

        muhit.harakat_qollash(harakat)

        print("Xonalar holati:", muhit.xonalar)
        print("Agent joylashuvi:", muhit.agent_joylashuvi)
        print("--------------------")


simulyatsiya(20)
```

(agent_joylashuvi, xonaning holati)
('A', 'iflos')
2 qoida:
('A', 'iflos')
if holat == "iflos":
    Tozalamoq
('A', 'toza')
if holat == "toza" and joy == "A":
    O'ngga yurish
('B', 'toza')
if holat == "toza" and joy == "B":
    Chapga yurish
3 action:
Agar xona iflos bo‚Äòlsa: Maqsad tozalikni saqlash. Shuning uchun tozalash tanlandi.
Agar xona toza bo‚Äòlsa: Agent boshqa xonani tekshirishi kerak. Shuning uchun harakat qiladi.

# **3-MASALA: Aqlli sug‚Äòorish agentini yaratish**

---


## **Maqsad:**

## Agentga mantiqiy va (and) operatori hamda inkor (not) tushunchasini qo‚Äòllash.

## **Vazifa:**

## Tuproq namligi va yomg‚Äòir bor-yo‚Äòqligiga qarab suv nasosini boshqarish.

* ### yomgir = True ‚Üí **NASOSNI_OCHIR**

* ### namlik < 30 va yomgir = False ‚Üí **NASOSNI_YOQ**

* ###  namlik >= 30 va yomgir == False ‚Üí NASOS_OCH ‚Üí **NASOSNI_OCHIR**
* ### Aks holda ‚Üí **TEKSHIRISH_REJIMI**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>namlik</code> (0-100 oralig'idagi son)
* ### <code>yomgir</code> (True yoki False)



##***Klasslar:***

* ### <code>BogMuhiti</code>

* ### <code>SugorishAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas


 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.


```python
# Kodni shu yerga yozing

import random

class BogMuhiti:
    def __init__(self):
        self.namlik = random.randint(0, 100)
        self.yomgir = random.choice([True, False])

    def perseptsiya_ol(self):
        return (self.namlik, self.yomgir)

    def harakat_qollash(self, harakat):
        # Keyingi qadam uchun yangi sharoit hosil qilamiz
        self.namlik = random.randint(0, 100)
        self.yomgir = random.choice([True, False])

class SugorishAgenti:
    def qaror_qabul_qil(self, perseptsiya):
        namlik, yomgir = perseptsiya

        if yomgir == True:
            return "NASOSNI_OCHIR"

        elif namlik < 30 and yomgir == False:
            return "NASOSNI_YOQ"

        elif namlik >= 30 and yomgir == False:
            return "NASOSNI_OCHIR"

        else:
            return "TEKSHIRISH_REJIMI"

def simulyatsiya(qadamlar=20):
    muhit = BogMuhiti()
    agent = SugorishAgenti()

    for i in range(1, qadamlar + 1):
        perseptsiya = muhit.perseptsiya_ol()
        harakat = agent.qaror_qabul_qil(perseptsiya)

        print(f"Qadam {i}")
        print("Namlik:", perseptsiya[0])
        print("Yomgir:", perseptsiya[1])
        print("Action:", harakat)
        print("--------------------")

        muhit.harakat_qollash(harakat)


simulyatsiya(20)
```

# **4-MASALA: Xavfsizlik qorovuli agentini yaratish.**


---


## **Maqsad:**

## Agent ikki xil sensordan kelgan ma'lumotlarni birlashtirib qaror qabul qilish.

## **Vazifa:**

## Harakat sensori va eshik qulfi holatiga qarab signal chalish.

* ### harakat bor va qulf ochiq bo'lsa  ‚Üí **SIGNAL_CHALISH**

* ### harakat bor va qulf yopiq bo'lsa ‚Üí **SALOMLASHISH**

* ###  harakat yo'q ‚Üí **JIM_TURISH**
* ### Aks holda ‚Üí **KUZATISH**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>harakat</code> (True yoki False)
* ### <code>qulf_ochiq</code> (True yoki False)



##***Klasslar:***

* ### <code>UyMuhiti</code>

* ### <code>QorovulAgenti</code>


##***Simulyatsiya:***
* ### 20 tadan kam emas

 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.



```python
# Kodni shu yerga yozing
import random

class UyMuhiti:
    def __init__(self):
        self.harakat = random.choice([True, False])
        self.qulf_ochiq = random.choice([True, False])

    def perseptsiya_ol(self):
        return (self.harakat, self.qulf_ochiq)

    def harakat_qollash(self, harakat):
        # Keyingi qadam uchun yangi sensor qiymatlari
        self.harakat = random.choice([True, False])
        self.qulf_ochiq = random.choice([True, False])

class QorovulAgenti:
    def qaror_qabul_qil(self, perseptsiya):
        harakat, qulf_ochiq = perseptsiya

        if harakat == True and qulf_ochiq == True:
            return "SIGNAL_CHALISH"

        elif harakat == True and qulf_ochiq == False:
            return "SALOMLASHISH"

        elif harakat == False:
            return "JIM_TURISH"

        else:
            return "KUZATISH"

def simulyatsiya(qadamlar=20):
    muhit = UyMuhiti()
    agent = QorovulAgenti()

    for i in range(1, qadamlar + 1):
        perseptsiya = muhit.perseptsiya_ol()
        harakat_natija = agent.qaror_qabul_qil(perseptsiya)

        print(f"Qadam {i}")
        print("Harakat sensori:", perseptsiya[0])
        print("Qulf ochiq:", perseptsiya[1])
        print("Action:", harakat_natija)
        print("--------------------")

        muhit.harakat_qollash(harakat_natija)


simulyatsiya(20)
```

agar (True, False)

harakat yo'q JIM_TURISH
3. action

# **5-MASALA: Liftni boshqaruvchi agentni yaratish.**

---


## **Maqsad:**

## Agent joriy holat va maqsadli holatni solishtiradi.

## **Vazifa:**

## Lift turgan qavat va chaqirilgan qavatga qarab harakat yo‚Äònalishini aniqlash.

* ### **YUQORIGA**
* ### **PASTGA**
* ### **ESHIKNI_OCH**
* ### Aks holda ‚Üí **KUTISH**


## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>lift_qavati</code> (1-9 oralig'ida)
* ### <code>chaqiruv_qavati</code> (1-9 oralig'ida)



##***Klasslar:***

* ### <code>BinoMuhiti</code>

* ### <code>LiftAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas


 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.

```python
# Kodni shu yerga yozing
import random

class BinoMuhiti:
    def __init__(self):
        self.lift_qavati = random.randint(1, 9)
        self.chaqiruv_qavati = random.randint(1, 9)

    def perseptsiya_ol(self):
        return (self.lift_qavati, self.chaqiruv_qavati)

    def harakat_qollash(self, harakat):
        # Keyingi qadam uchun yangi chaqiruv va lift holati
        self.lift_qavati = random.randint(1, 9)
        self.chaqiruv_qavati = random.randint(1, 9)

class LiftAgenti:
    def qaror_qabul_qil(self, perseptsiya):
        lift_qavati, chaqiruv_qavati = perseptsiya

        if lift_qavati < chaqiruv_qavati:
            return "YUQORIGA"

        elif lift_qavati > chaqiruv_qavati:
            return "PASTGA"

        elif lift_qavati == chaqiruv_qavati:
            return "ESHIKNI_OCH"

        else:
            return "KUTISH"

def simulyatsiya(qadamlar=20):
    muhit = BinoMuhiti()
    agent = LiftAgenti()

    for i in range(1, qadamlar + 1):
        perseptsiya = muhit.perseptsiya_ol()
        harakat = agent.qaror_qabul_qil(perseptsiya)

        print(f"Qadam {i}")
        print("Lift qavati:", perseptsiya[0])
        print("Chaqiruv qavati:", perseptsiya[1])
        print("Action:", harakat)
        print("--------------------")

        muhit.harakat_qollash(harakat)


simulyatsiya(20)
```

1. percept
(lift_qavati, chaqiruv_qavati)
masalan (3, 7)
Bu degani lift 3-qavatda 7-qavatdan chaqiruv bor
2. qoida
agar (3, 7)
lift_qavati < chaqiruv_qavati YUQORIGA
agar (8, 2)
lift_qavati > chaqiruv_qavati PASTGA
agar (5, 5)
lift_qavati == chaqiruv_qavati ESHIKNI_OCH
3. action
Agar lift pastda va chaqiruv yuqorida ‚Üí yuqoriga harakat qiladi.
Agar lift yuqorida va chaqiruv pastda ‚Üí pastga harakat qiladi.
Agar lift chaqirilgan qavatda bo‚Äòlsa ‚Üí eshikni ochadi.
Aks holda ‚Üí kutish (formal to‚Äòliq qamrov uchun).

# **6-MASALA: Server yuklamasini balanslovchi agent yaratish.**

---

## **Maqsad:**

## Agent resurslarni boshqarish uchun chegaraviy qiymatlar (thresholds) bilan ishlaydi.

## **Vazifa:**

## Protsessor (CPU) yuklamasiga qarab qo‚Äòshimcha serverlarni yoqish yoki o‚Äòchirish.

* ### **SERVER_QOSHISH**
* ### **SERVER_OCHIRISH**
* ### **STABIL_HOLAT**


## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>cpu_yuklama</code> (0-100 foiz oralig'ida)

##***Klasslar:***

* ### <code>DataMarkazMuhiti</code>

* ### <code>BalansAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas

 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.

```python
# Kodni shu yerga yozing

import random

class DataMarkazMuhiti:
    def __init__(self):
        self.cpu_yuklama = random.randint(0, 100)

    def perseptsiya_ol(self):
        return self.cpu_yuklama

    def harakat_qollash(self, harakat):
        # Keyingi qadam uchun yangi yuklama qiymati
        self.cpu_yuklama = random.randint(0, 100)

class BalansAgenti:
    def qaror_qabul_qil(self, cpu_yuklama):

        if cpu_yuklama > 75:
            return "SERVER_QOSHISH"

        elif cpu_yuklama < 30:
            return "SERVER_OCHIRISH"

        else:
            return "STABIL_HOLAT"

def simulyatsiya(qadamlar=20):
    muhit = DataMarkazMuhiti()
    agent = BalansAgenti()

    for i in range(1, qadamlar + 1):
        cpu = muhit.perseptsiya_ol()
        harakat = agent.qaror_qabul_qil(cpu)

        print(f"Qadam {i}")
        print("CPU yuklama:", cpu, "%")
        print("Action:", harakat)
        print("--------------------")

        muhit.harakat_qollash(harakat)


simulyatsiya(20)
```

cpu_yuklama
Masalan: 82
Bu degani:
CPU 82% band
2. qoida
Agar percept: 82
cpu_yuklama > 75 SERVER_QOSHISH
Agar percept: 20
cpu_yuklama < 30 SERVER_OCHIRISH
Agar percept:55
else STABIL_HOLAT
3. action
Agar CPU yuklama juda yuqori bo‚Äòlsa ‚Üí tizim zo‚Äòriqmoqda ‚Üí qo‚Äòshimcha server qo‚Äòshiladi.
Agar yuklama juda past bo‚Äòlsa ‚Üí resurs isrof ‚Üí server o‚Äòchiriladi.
Agar yuklama normal oraliqda bo‚Äòlsa ‚Üí tizim stabil ‚Üí hech narsa o‚Äòzgarmaydi.

# **8-MASALA: Spam aniqlovchi agent yaratish.**

---

## **Maqsad:**

## Matnli ma'lumotlar (String) ichidan kalit so‚Äòzlarni qidirish va saralashni o‚Äòrganish.

## **Vazifa:**

## Kelgan xabarning tarkibida shubhali so‚Äòzlar bor-yo‚Äòqligiga qarab uni tegishli papkaga joylashtirish.

* ### xabar tarkibida "yutuq" yoki "tekin" so‚Äòzlari qatnashsa ‚Üí **SPAM_PAPKA**
* ### xabar tarkibida "hisobot" yoki "ish" so‚Äòzlari qatnashsa ‚Üí **MUHIM_PAPKA**
* Aks holda ‚Üí **KIRUVCHI_PAPKA**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>xabar</code> (String - matn ko‚Äòrinishida)

##***Klasslar:***

* ### <code>PochtaMuhiti</code>

* ### <code>SpamAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas

 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.


```python
# Kodni shu yerga yozing

import random

class PochtaMuhiti:
    def __init__(self):
        self.xabarlar = [
            "Siz yutuq yutdingiz",
            "Bugun tekin sovg'a bor",
            "Oylik hisobotni yuboring",
            "Yangi ish taklifi",
            "Salom, qalaysiz?"
        ]
        self.xabar = random.choice(self.xabarlar)

    def perseptsiya_ol(self):
        return self.xabar

    def harakat_qollash(self, harakat):
        # Keyingi qadam uchun yangi xabar
        self.xabar = random.choice(self.xabarlar)

class SpamAgenti:
    def qaror_qabul_qil(self, xabar):
        matn = xabar.lower()

        if "yutuq" in matn or "tekin" in matn:
            return "SPAM_PAPKA"

        elif "hisobot" in matn or "ish" in matn:
            return "MUHIM_PAPKA"

        else:
            return "KIRUVCHI_PAPKA"

def simulyatsiya(qadamlar=20):
    muhit = PochtaMuhiti()
    agent = SpamAgenti()

    for i in range(1, qadamlar + 1):
        xabar = muhit.perseptsiya_ol()
        harakat = agent.qaror_qabul_qil(xabar)

        print(f"Qadam {i}")
        print("Xabar:", xabar)
        print("Action:", harakat)
        print("--------------------")

        muhit.harakat_qollash(harakat)


simulyatsiya(20)
```

xabar (String)

2. qoida
agar: "Siz yutuq yutdingiz"
"yutuq" mavjud SPAM_PAPKA
agar: "Oylik hisobotni yuboring"
"hisobot" mavjud MUHIM_PAPKA
agar: "Salom, qalaysiz?"
else ‚Üí KIRUVCHI_PAPKA
3. action
Agar matnda "yutuq" yoki "tekin" bo‚Äòlsa reklama yoki spam ehtimoli SPAM_PAPKA.
Agar "hisobot" yoki "ish" bo`lsa ishga oid muhim xabar MUHIM_PAPKA.
Boshqa hollarda oddiy kiruvchi xabar.

## **9-MASALA: Aqlli avtomobil parkovka agentini yaratish.**

---

## **Maqsad:**

## Masofa datchiklari (sensorlari) asosida obyektgacha bo‚Äòlgan oraliqni aniqlash va xavfsiz to‚Äòxtashni ta‚Äôminlash.

## **Vazifa:**

## Avtomobil orqaga harakatlanayotganda devorgacha bo‚Äòlgan masofaga qarab tormozni boshqarish.

* ### masofa <= 0.5 ‚Üí **TOXTASH** (Avariya xavfi)

* ### masofa > 0.5 va masofa <= 2.0 ‚Üí **SEKINLASHISH** (Ogohlantirish signali)

* ### Aks holda ‚Üí **ORQAGA_YURISH**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>masofa</code> (float - metr hisobida, masalan: 5.4, 1.2, 0.3)

##***Klasslar:***

* ### <code>GarajMuhiti</code>

* ### <code>ParkovkaAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas

 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.


```python
# Kodni shu yerga yozing

import random

class GarajMuhiti:
    def __init__(self):
        self.masofa = round(random.uniform(0.0, 5.5), 2)

    def perseptsiya_ol(self):
        return self.masofa

    def harakat_qollash(self, harakat):
        # Keyingi qadam uchun yangi masofa qiymati
        self.masofa = round(random.uniform(0.0, 5.5), 2)

class ParkovkaAgenti:
    def qaror_qabul_qil(self, masofa):

        if masofa <= 0.5:
            return "TOXTASH"

        elif masofa > 0.5 and masofa <= 2.0:
            return "SEKINLASHISH"

        else:
            return "ORQAGA_YURISH"

def simulyatsiya(qadamlar=20):
    muhit = GarajMuhiti()
    agent = ParkovkaAgenti()

    for i in range(1, qadamlar + 1):
        masofa = muhit.perseptsiya_ol()
        harakat = agent.qaror_qabul_qil(masofa)

        print(f"Qadam {i}")
        print("Masofa:", masofa, "metr")
        print("Action:", harakat)
        print("--------------------")

        muhit.harakat_qollash(harakat)


simulyatsiya(20)
```

1. percept
masofa (float)
Masalan: 1.2
Bu degani avtomobil devordan 1.2 metr uzoqlikda
2. qoida
Agar 0.3
masofa <= 0.5 TOXTASH
Agar 1.4
0.5 < masofa <= 2.0 SEKINLASHISH
Agar 4.8
else ORQAGA_YURISH
3. action
Agar masofa 0.5 metr yoki undan kam bo‚Äòlsa avariya xavfi darhol to‚Äòxtash.
Agar masofa 0.5‚Äì2 metr oralig‚Äòida bo‚Äòlsa xavf yaqinlashmoqda sekinlashish.
Agar masofa 2 metrdan katta bo‚Äòlsa xavfsiz zona orqaga yurishda davom etadi.

## **10-MASALA: Aqlli muzlatkich agentini yaratish.**

---

## **Maqsad:**

## Bir nechta (3 ta) sensordan kelgan ma'lumotlarni ustuvorlik bo‚Äòyicha tahlil qilish.

## **Vazifa:**

## Muzlatkich eshigi, ichki harorat va mahsulot miqdoriga qarab, mos xabarni chiqarish.

* ### **SIGNAL_CHALISH** (Eng muhimi)
* ### **SOVITISHNI_KUCHAYTIRISH**
* ### **NORMAL_REJIM**
* ### Aks holda ‚Üí **BUYURTMA_BERISH**
* ### Aks holda ‚Üí **NORMAL_HOLAT**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>eshik_ochiq</code> (True yoki False)

* ### <code>harorat</code> (butun son, masalan: 4, 8, -2)

* ### <code>mahsulot_soni</code> (butun son)

##***Klasslar:***

* ### <code>OshxonaMuhiti</code>

* ### <code>MuzlatkichAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas

## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.


```python
# Kodni shu yerga yozing

import random

class OshxonaMuhiti:
    def __init__(self):
        self.eshik_ochiq = random.choice([True, False])
        self.harorat = random.randint(-5, 10)
        self.mahsulot_soni = random.randint(0, 20)

    def perseptsiya_ol(self):
        return (self.eshik_ochiq, self.harorat, self.mahsulot_soni)

    def harakat_qollash(self, harakat):
        # Keyingi qadam uchun yangi qiymatlar
        self.eshik_ochiq = random.choice([True, False])
        self.harorat = random.randint(-5, 10)
        self.mahsulot_soni = random.randint(0, 20)

class MuzlatkichAgenti:
    def qaror_qabul_qil(self, perseptsiya):
        eshik_ochiq, harorat, mahsulot_soni = perseptsiya

        # 1-ustuvorlik
        if eshik_ochiq == True:
            return "SIGNAL_CHALISH"

        # 2-ustuvorlik
        elif harorat > 6:
            return "SOVITISHNI_KUCHAYTIRISH"

        # 3-ustuvorlik
        elif mahsulot_soni < 3:
            return "BUYURTMA_BERISH"

        # 4-holat
        else:
            return "NORMAL_HOLAT"

def simulyatsiya(qadamlar=20):
    muhit = OshxonaMuhiti()
    agent = MuzlatkichAgenti()

    for i in range(1, qadamlar + 1):
        perseptsiya = muhit.perseptsiya_ol()
        harakat = agent.qaror_qabul_qil(perseptsiya)

        print(f"Qadam {i}")
        print("Eshik ochiq:", perseptsiya[0])
        print("Harorat:", perseptsiya[1])
        print("Mahsulot soni:", perseptsiya[2])
        print("Action:", harakat)
        print("--------------------")

        muhit.harakat_qollash(harakat)


simulyatsiya(20)
```

1. percept
(eshik_ochiq, harorat, mahsulot_soni)
Masalan:(True, 5, 10)
Bu degani eshik ochiq ichki harorat 5¬∞ mahsulot 10 ta
2. qoida
agar (True, 3, 8)
eshik_ochiq == True SIGNAL_CHALISH
Agar (False, 8, 10)
harorat > 6 SOVITISHNI_KUCHAYTIRISH
Agar (False, 3, 1)
mahsulot_soni < 3 BUYURTMA_BERISH
Agar (False, 3, 8)
else ‚Üí NORMAL_HOLAT
3. action
Eshik ochiq bo‚Äòlsa sovuq chiqib ketadi darhol signal (eng muhim).
Harorat yuqori bo‚Äòlsa ‚Üí sovitishni kuchaytirish kerak.
Mahsulot kam bo‚Äòlsa ‚Üí buyurtma berish kerak.
Aks holda ‚Üí tizim normal ishlamoqda.
